```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
FES=np.genfromtxt('fes.dat')
```

```python
cv=FES[:,0]
fes=FES[:,1]
```

```python
plt.plot(cv,fes,'b')
plt.xlabel('RMSD (nm)')
plt.ylabel('Free Energy (kJ/mol)')
plt.title('Free Energy vs RMSD for Chignolin protein')
plt.savefig('FES_chignolin.png',dpi=300,bbox_inches='tight')
```
